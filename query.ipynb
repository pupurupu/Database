```python
import numpy as np
import pandas as pd
import lightkurve as lk
import matplotlib.pyplot as plt
import ztfquery
from ztfquery import lightcurve
from astropy.coordinates import SkyCoord
from astropy.coordinates import Angle
import astropy.coordinates as coord
from astroquery.vizier import Vizier
from astropy.units import u
import astropy.units as u
```

```python
dat = pd.read_table("subset_9.txt", header = 0, sep = ',')

vizier = Vizier(columns=['**'])
```

```python
dat
```

```python
Gaia_tess = []
Bmag_tess = []
Vmag_tess = []
Gmag_tess = []
Tmag_tess = []

tess = []

for i in range(len(dat)):
    tess.append(vizier.query_region(coord.SkyCoord(ra=dat['ra'][i], dec=dat['dec'][i], unit=(u.deg,u.deg), 
                                                   frame='icrs'), radius = Angle(0.001, 'deg'), catalog=['TESS', 'IV/39/']))

for i in range(len(dat)):
    if bool(tess[i])==True: 
        Gaia_tess.append(np.ma.getdata(tess[i][0]['GAIA'])[0])
        Bmag_tess.append(np.ma.getdata(tess[i][0]['Bmag'])[0])
        Vmag_tess.append(np.ma.getdata(tess[i][0]['Vmag'])[0])
        Gmag_tess.append(np.ma.getdata(tess[i][0]['Gmag'])[0])
        Tmag_tess.append(np.ma.getdata(tess[i][0]['Tmag'])[0])
    
    else:
        Gaia_tess.append(['nan'][0])
        Bmag_tess.append(['nan'][0])
        Vmag_tess.append(['nan'][0])
        Gmag_tess.append(['nan'][0])
        Tmag_tess.append(['nan'][0])

dat['GAIA'] = Gaia_tess
dat['Bmag_t'] = Bmag_tess
dat['Vmag_t'] = Vmag_tess
dat['Gmag_t'] = Gmag_tess
dat['Tmag_t'] = Tmag_tess
```

```python
Bmag_nomad = []
Vmag_nomad = []
Rmag_nomad = []
Jmag_nomad = []
Hmag_nomad = []
Kmag_nomad = []

result = []
for i in range(len(dat)):
    result.append(vizier.query_region(coord.SkyCoord(ra=dat['ra'][i], dec=dat['dec'][i], unit=(u.deg,u.deg), 
                                                     frame='icrs'), radius = Angle(0.001, 'deg'), catalog=['NOMAD', 'I/297']))

for i in range(len(dat)):
    if bool(result[i])==True: 
        Bmag_nomad.append(np.ma.getdata(result[i][0]['Bmag'])[0])
        Vmag_nomad.append(np.ma.getdata(result[i][0]['Vmag'])[0])
        Rmag_nomad.append(np.ma.getdata(result[i][0]['Rmag'])[0])
        Jmag_nomad.append(np.ma.getdata(result[i][0]['Jmag'])[0])
        Hmag_nomad.append(np.ma.getdata(result[i][0]['Hmag'])[0])
        Kmag_nomad.append(np.ma.getdata(result[i][0]['Kmag'])[0])
    else:
        Bmag_nomad.append(['nan'][0])
        Vmag_nomad.append(['nan'][0])
        Rmag_nomad.append(['nan'][0])
        Jmag_nomad.append(['nan'][0])
        Hmag_nomad.append(['nan'][0])
        Kmag_nomad.append(['nan'][0])
      
dat['Bmag_N'] = Bmag_nomad
dat['Vmag_N'] = Vmag_nomad
dat['Rmag_N'] = Rmag_nomad
dat['Jmag_N'] = Jmag_nomad
dat['Hmag_N'] = Hmag_nomad
dat['Kmag_N'] = Kmag_nomad
```

```python
Bmag_aavso = []
Vmag_aavso = []


aavso = []
for i in range(len(dat)):
    aavso.append(vizier.query_region(coord.SkyCoord(ra=dat['ra'][i], dec=dat['dec'][i], 
                                            unit=(u.deg,u.deg), frame='icrs'), 
                             radius = Angle(0.001, 'deg'), catalog=['AAVSO', 'II/336']))

for i in range(len(dat)):
    if bool(aavso[i])==True: 
        Bmag_aavso.append(np.ma.getdata(aavso[i][0]['Bmag'])[0])
        Vmag_aavso.append(np.ma.getdata(aavso[i][0]['Vmag'])[0])
    
    else:
        Bmag_aavso.append(['nan'][0])
        Vmag_aavso.append(['nan'][0])
    
dat['Bmag_A'] = Bmag_aavso
dat['Vmag_A'] = Vmag_aavso
```

```python
Jmag_2mass = []
Hmag_2mass = []
Kmag_2mass = []

mass = []
for i in range(len(dat)):
    mass.append(vizier.query_region(coord.SkyCoord(ra=dat['ra'][i], dec=dat['dec'][i], 
                                            unit=(u.deg,u.deg), frame='icrs'), 
                             radius = Angle(0.001, 'deg'), catalog=['2MASS', 'II/246']))

for i in range(len(dat)):
    if bool(mass[i])==True: 
        Jmag_2mass.append(np.ma.getdata(mass[i][0]['Jmag'])[0])
        Hmag_2mass.append(np.ma.getdata(mass[i][0]['Hmag'])[0])
        Kmag_2mass.append(np.ma.getdata(mass[i][0]['Kmag'])[0])
    else:
        Jmag_2mass.append(['nan'][0])
        Hmag_2mass.append(['nan'][0])
        Kmag_2mass.append(['nan'][0])

dat['Jmag_2M'] = Jmag_2mass
dat['Hmag_2M'] = Hmag_2mass
dat['Kmag_2M'] = Kmag_2mass
```

```python
dat.to_csv('TyulisP_06-201.csv', header=True, index=False)
```
